### Years 2016, 2017, 2018, 2019

---

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
import time
```

```python
h16 = pd.read_csv('./data/half_2016.csv', usecols=['Name', 'Time'])
h17 = pd.read_csv('./data/half_2017.csv', usecols=['Name', 'Time'])
h18 = pd.read_csv('./data/half_2018.csv', usecols=['Name', 'Time'])
h19 = pd.read_csv('./data/half_2019.csv', usecols=['Name', 'Time'])
```

---

### Convert Times and Clean Split up non-Finishers

 - `h16_non` = Marathoners who did NOT finish
 - `h16_fin` = Marathoners who finished

```python
h16['Seconds'] = h16['Time'].apply(lambda x : pd.to_timedelta(x).total_seconds())
h17['Seconds'] = h17['Time'].apply(lambda x : pd.to_timedelta(x).total_seconds())
h18['Seconds'] = h18['Time'].apply(lambda x : pd.to_timedelta(x).total_seconds())
h19['Seconds'] = h19['Time'].apply(lambda x : pd.to_timedelta(x).total_seconds())
```

```python
h16_non = h16[h16['Seconds'] == 0.0]
h16_fin = h16[h16['Seconds'] != 0.0]

h17_non = h17[h17['Seconds'] == 0.0]
h17_fin = h17[h17['Seconds'] != 0.0]

h18_non = h18[h18['Seconds'] == 0.0]
h18_fin = h18[h18['Seconds'] != 0.0]

h19_non = h19[h19['Seconds'] == 0.0]
h19_fin = h19[h19['Seconds'] != 0.0]
```

---

## HALF Marathon Plotting

```python
r = ['16`','17`','18`','19`']
years_fin = [len(h16_fin),len(h17_fin),len(h18_fin),len(h19_fin)] 
years_non_fin = [len(h16_non),len(h17_non),len(h18_non),len(h19_non)]


fig, ax = plt.subplots(figsize=(18, 8));

ax1 = plt.bar(r, years_fin, 
        color='lightblue', 
        edgecolor='white', 
        width=.8,
        label='finishers'
        );
ax2 = plt.bar(r, 
        years_non_fin, 
        bottom=years_fin, 
        color='salmon', 
        edgecolor='white', 
        width=.8,
        label='non-finishers'
        );


plt.title('HALF Marathoners Per Year\n'.upper(), fontsize=20);
plt.yticks([0,13000, 25000]);
plt.xticks(r, fontweight='bold');
plt.legend(handles=[ax1, ax2]);

plt.savefig('./img_data/half_fin_non-fin_all_years');
```
